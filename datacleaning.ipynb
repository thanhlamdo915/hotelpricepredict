```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = pd.read_csv('hoteldata.csv', on_bad_lines='skip')


```

```python
data1.head()
```

```python
data1
```

```python
data=data1
data.columns
```

```python
data.info()
```

```python
cols_to_drop = [' nearest',' distance nearest','tax.1','Unnamed: 0']
data_clean = data.drop(cols_to_drop, axis=1)
data_clean.rename(columns={" location": "location"}, inplace = True)
```

```python
data_clean.info()
```

```python
data_clean
```

```python
mean_value_rating = data_clean['rating'].mean()
mean_value_reviews = data_clean['reviews'].mean()

data_clean['rating'].fillna(value=mean_value_rating, inplace=True)
data_clean['reviews'].fillna(value=mean_value_reviews, inplace=True)
data_clean
```

```python
data_clean['reviews'] = data_clean['reviews'].round()

data_clean
```

```python
data_clean['tax'].fillna(value=0, inplace=True)
data_clean['tax'] = data_clean['tax'].replace(r'^\s*$', np.NaN, regex=True)
data_clean
```

```python

data_clean.dropna(subset='star rating', how='all', inplace=True)
data_clean
```

```python
data_clean['tax'].fillna(value=0, inplace=True)
```

```python
data_clean.info()
```

```python
result=data_clean.drop_duplicates()
```

```python
result.info()
```

```python
result['price'] = result['price'].str.split('D', expand=True)[1].astype(int)
```

```python
result
```

```python
result.reset_index(drop=True, inplace=True)
```

```python
result
```
